```python
# create and save dictionary of model paths
from glob import glob
models_dirs = glob('shapenet_chairs/*/')
model_dicts = {}
i = 0
for k, v in enumerate(models_dirs):
  test1 = os.path.isdir(models_dirs[k]+'images')
  test2 = os.path.isdir(models_dirs[k]+'models/untitled')
  if test1 or test2:
    model_dicts[i] = v
    i += 1
  else:
    print('No texture bound, discard. '+str(k)+' '+str(v))
import json
with open(os.getcwd()+'/model_maps.json', 'w') as f:
  json.dump(model_dicts, f)
f.close()
```
