```python
import numpy as np
import h5py
import pandas as pd
import cv2
```

```python
# reference: https://github.com/VVasanth/MachineLearning/tree/master/HDF5DataAug
# prepare paths
hdf5_path = os.getcwd()+'/data.hdf5'
index_df = pd.read_csv(os.getcwd()+'/data_paths.csv')
raw_image_paths = index_df['raw_image_file_paths']
normal_masked_paths = index_df['normal_masked_paths']
depth_masked_paths = index_df['depth_masked_paths']
```

```python
pair1 = list(zip(raw_image_paths, normal_masked_paths))
pair2 = list(zip(raw_image_paths, depth_masked_paths))
```

```python
depth_gray_shape = (len(index_df.index), 256, 256, 1)
depth_rgb_shape = (len(index_df.index), 256, 256, 3)
normal_shape = (len(index_df.index), 256, 256, 3)
raw_image_shape = (len(index_df.index), 256, 256, 3)

f = h5py.File(hdf5_path, mode='w')

f.create_dataset('depth', depth_gray_shape, np.float32, compression='lzf')
# f.create_dataset('depth_rgb', depth_rgb_shape, np.float32, compression='lzf')
f.create_dataset('normal', normal_shape, np.float32, compression='lzf')
f.create_dataset('raw_image', raw_image_shape, np.float32, compression='lzf')
```

```python
for i in range(len(index_df.index)):
  if i % 100 == 0 and i > 1:
    print('Depth: {}/{}'.format(i, depth_gray_shape[0]))
  addr = depth_masked_paths[i]
  img = cv2.imread(addr)
  img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img = np.atleast_3d(img)
  f['depth'][i, ...] = img[None]

# for i in range(len(index_df.index)):
#   if i % 10000 == 0 and i > 1:
#     print('Depth: {}/{}'.format(i, depth_rgb_shape[0]))
#   addr = depth_masked_paths[i]
#   img = cv2.imread(addr)
#   img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
#   img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#   f['depth_rgb'][i, ...] = img[None]

for i in range(len(index_df.index)):
  if i % 10000 == 0 and i > 1:
    print('Normal: {}/{}'.format(i, normal_shape[0]))
  addr = normal_masked_paths[i]
  img = cv2.imread(addr)
  img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  f['normal'][i, ...] = img[None]

for i in range(len(index_df.index)):
  if i % 10000 == 0 and i > 1:
    print('Raw image: {}/{}'.format(i, raw_image_shape[0]))
  addr = raw_image_paths[i]
  img = cv2.imread(addr)
  img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  f['raw_image'][i, ...] = img[None]
```
