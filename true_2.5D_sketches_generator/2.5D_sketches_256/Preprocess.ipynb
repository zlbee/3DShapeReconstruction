```python
# sort files and move them into sub-directories
def move_to_subdir(directory_name, file_name):
    model_paths_total = []
    for model_idx in range(4298):
        model_paths = []
        for view in range(10):
            model_path = './'+directory_name+'/'+str(model_idx)+'_'+str(view)+'_'+file_name+'.jpg'
            model_paths.append(model_path)
        model_paths_total.append(model_paths)

    for model_idx, model_paths in enumerate(model_paths_total):
        new_path = './'+directory_name+'/' + str(model_idx)
        os.mkdir(new_path)
        for model_path in model_paths:
            shutil.move(model_path, new_path)
            print(model_path+' moved')
```
