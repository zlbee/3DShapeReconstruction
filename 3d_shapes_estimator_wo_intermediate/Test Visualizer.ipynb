```python
import os

import numpy as np
import matplotlib.pyplot as plt

def voxels_to_coordiates(voxels):
  coordinates = []
  for x in range(128):
    for y in range(128):
      for z in range(128):
        if voxels[x][y][z] == 1:
          coordinates.append([x, y, z])
  coordinates = np.asarray(coordinates)
  return coordinates

# view samples
dir = OUTPUT_PATH
for j in range(1, 8):
  for i in range(len(os.listdir(dir))-1):
    # img raw
    addr = dir+'/real_raw_image%02d.npz' % i
    img0 = np.load(addr)[j].astype(np.float32)

    # img depth
    addr = dir+'/real_depth_image%02d.npz' % i
    img1 = np.load(addr)[j].astype(np.float32)

    # img normal
    addr = dir+'/real_normal_image%02d.npz' % i
    img2 = np.load(addr)[j].astype(np.float32)

    # predicted
    addr = dir+'/predicted_shape%02d.npz' % i
    shape = np.load(addr)[j][:, :, :, 0]
    shape = np.where(shape > 0.5, 1, 0).astype(int)
    coords1 = voxels_to_coordiates(shape)

    # real
    addr = dir+'/real_shape%02d.npz' % i
    shape = np.load(addr)[j][:, :, :, 0]
    shape = np.where(shape > 0.5, 1, 0).astype(int)
    coords2 = voxels_to_coordiates(shape)

    # fig = plt.figure(figsize=plt.figaspect(0.5))
    fig = plt.figure(figsize=(25, 5))


    # plot real sketch
    ax = fig.add_subplot(1, 5, 1)
    ax.imshow(img0)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # plot predicted sketch
    ax = fig.add_subplot(1, 5, 2)
    ax.imshow(img1)
    ax.set_xticks([])
    ax.set_yticks([])

    # plot predicted sketch
    ax = fig.add_subplot(1, 5, 3)
    ax.imshow(img2)
    ax.set_xticks([])
    ax.set_yticks([])

    # plot real shape
    ax = fig.add_subplot(1, 5, 4, projection='3d')
    ax.scatter(coords2[:, 0], coords2[:, 2], coords2[:, 1], alpha=0.1, c=coords2[:, 1])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_xticks(np.arange(0, 128, 20))
    ax.set_yticks(np.arange(0, 128, 20))
    ax.set_zticks(np.arange(0, 128, 20))
    ax.set_xlim(0, 128)
    ax.set_ylim(0, 128)
    ax.set_zlim(0, 128)
    ax.set_axis_off()

    # plot predicted shape
    ax = fig.add_subplot(1, 5, 5, projection='3d')
    ax.scatter(coords1[:, 0], coords1[:, 2], coords1[:, 1], alpha=0.1, c=coords1[:, 1])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_xticks(np.arange(0, 128, 20))
    ax.set_yticks(np.arange(0, 128, 20))
    ax.set_zticks(np.arange(0, 128, 20))
    ax.set_xlim(0, 128)
    ax.set_ylim(0, 128)
    ax.set_zlim(0, 128)
    ax.set_axis_off()

    isExist = os.path.exists(OUTPUT_PATH+'/visualization')
    if not isExist:
      # Create a new directory because it does not exist 
      os.makedirs(OUTPUT_PATH+'/visualization')
      print("The new directory is created!")

    # plt.savefig(OUTPUT_PATH+'/visualization/fig'+str(i)+'.jpg')
    plt.show()

pass
```
